```python
# importamos las librerias necesarisas:

import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
from scipy.optimize import minimize

# 1.- Derivación simbólica con Sympy

# definimos la variable simbólica
x = sp.Symbol('x')

# definimos la función f(x) = (x - 3)^2
f_expr = (x - 3)**2

# calculamos la derivada simbolica f'(x)
f_derivada = sp.diff(f_expr, x)

# igualamos f'(x) = 0 para encontrar puntos críticos
punto_critico = sp.solve(f_derivada, x)

print("Derivada simbólica f'(x):", f_derivada)
print("Punto crítico encontrado:", punto_critico)


# 2.- Visualización de f(x) y f'(x)

# convertimos las funciones simbólicas a funciones de Python
f_lambda = sp.lambdify(x, f_expr, modules=['numpy'])
f_deriv_lambda = sp.lambdify(x, f_derivada, modules=['numpy'])

# generamos valores en el intervalo [-5, 10]
x_vals = np.linspace(-5, 10, 300)
y_vals = f_lambda(x_vals)
y_deriv_vals = f_deriv_lambda(x_vals)

# graficamos f(x)
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(x_vals, y_vals, label='f(x) = (x - 3)^2')
plt.scatter([3], [0], color='red', label='mínimo en x=3')
plt.title('Función f(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)

# graficamos f'(x)
plt.subplot(1, 2, 2)
plt.plot(x_vals, y_deriv_vals, label="f'(x)", color='orange')
plt.axhline(0, color='black', linestyle='--', linewidth=0.8)
plt.scatter([3], [0], color='red', label='f\'(x) = 0 en x=3')
plt.title('Derivada f\'(x)')
plt.xlabel('x')
plt.ylabel('f\'(x)')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# 3.- Optimización Numérica con SciPy

# definimos f(x) como función numérica para minimizar
def f_numerica(x):
    return (x - 3)**2

# usamos minimize con un valor inicial arbitrario
resultado = minimize(f_numerica, x0=np.array([0]))

print("Resultado numérico con scipy.optimize.minimize:")
print("x que minimiza f(x):", resultado.x)
print("Valor mínimo encontrado:", resultado.fun)

# finalmente evaluamos el algoritmo mediante la comprobación de coincidencia con la solución simbólica:
if np.isclose(resultado.x[0], float(punto_critico[0]), atol=1e-5):
    print("Coincidencia: el mínimo encontrado numéricamente es el mismo que el punto crítico simbólico. ¡El algoritmo es confiable!")
else:
    print("Diferencia detectada entre optimización numérica y simbólica.")
```

Se definió la función \( f(x) = (x - 3)^2 \) y se utilizó la librería `SymPy` para calcular su derivada simbólica. A partir de esta derivada se resolvió la ecuación \( f'(x) = 0 \), obteniéndose el punto crítico \( x = 3 \), el cual corresponde al mínimo de la función.

Este procedimiento permitió validar teóricamente la ubicación del mínimo sin recurrir a evaluación numérica.

## Visualización de la función y su derivada

Se generaron valores de \( x \) en el intervalo \([-5, 10]\) y se utilizaron las funciones derivadas de `SymPy` para obtener los valores correspondientes de \( f(x) \) y \( f'(x) \).

Posteriormente, se graficaron ambas funciones utilizando `Matplotlib`, marcando de forma destacada el punto crítico. La visualización permitió observar que en \( x = 3 \) la pendiente de la curva es nula y la función alcanza su valor mínimo.

## Optimización numérica

Se empleó la función `minimize` de la biblioteca `scipy.optimize` para encontrar el valor de \( x \) que minimiza la función \( f(x) \). El método numérico comenzó desde una estimación inicial y utilizó procedimientos iterativos para aproximarse al mínimo.

El resultado obtenido fue \( x \approx 3 \), coincidiendo con el punto crítico hallado mediante derivación simbólica. Esta coincidencia valida la consistencia entre los enfoques simbólico y numérico.
